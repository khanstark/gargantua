```
import requests 
import threading
from bs4 import BeautifulSoup 
import operator 
from collections import Counter,defaultdict 
primary_keyword=["nitk"]

# Enter search keywords in this list
wordlist = ["conference","ananthanarayana","qwerty","Entrepreneurs"]
out=defaultdict(list)
url = "http://www.infotech.nitk.ac.in"
stack=[url]
links=[]
temp=stack
wordlist=primary_keyword + wordlist

def starts(): 

    while(len(stack)!=0):
        baseURL=stack.pop()
        if baseURL.split("/")[2]=="www.infotech.nitk.ac.in": 
            try:
                source_code = requests.get(baseURL).text
                soup = BeautifulSoup(source_code, 'html.parser') 
                data=str(soup.body.text)
                curr_page_words = data.lower().split()
        
                #print(curr_page_words)
                print(stack)
                for searched_word in wordlist:
                    if searched_word.lower() in curr_page_words:
                        if baseURL not in out[searched_word]: 
                            out[searched_word].append(baseURL)
                        else:
                            continue
                        #print(out)
                        print(searched_word)
                        for link in soup.find_all('a'):
                            href=str(link.get('href'))
                            links.append(href)
                            if not (href.startswith('http') and href.startswith('#main') and href.startswith('javascript') and href=="None") :
                                if len(stack)<=70:
                                    stack.append(url+href)
                                    temp.append(url+href)
                            else:
                                stack.append(href)
                                temp.append(href)
                    
            except:
                pass
          
    
threads = []
for i in range(3):
    t = threading.Thread(target=starts,name=i)
    threads.append(t)
    t.start()
for thread in threads:
        thread.join()
print(out)


#start("https://kaifkhan.000webhostapp.com/")
  
```

```
from IPython.display import display,HTML,IFrame
import operator
from collections import OrderedDict as od

length=len(out.keys())
sorted_dict=od()
__temp={}
__temp1=od()
for k,v in out.items():
  __temp[k]=len(v)

__temp1=od(sorted(__temp.items(),key=operator.itemgetter(1),reverse=True))
for k,v in __temp1.items():
  sorted_dict[k]=out[k]

#font="courier"
__global="<h1 align='center' style='color:#3b5998'><i>Search Results</i></h1><hr width='20%'/><p>Search successful for : "+str(length)+" keywords"
for k,v in sorted_dict.items():
  s="<br><h3>"+k+"</h3><br>"
  s+="<ul>"
  for value in v:
    s+="<li><a href="+value+">"+value+"</a></li>"
  s+="</ul>"
  __global+=s
  
not_present=[]
for word in wordlist:
  if word not in sorted_dict.keys():
    not_present.append(word)
    
for word in not_present:
  s="<br><h3>"+word+"</h3><br>"
  s+="<ul><li style='color:gray'>Not found in domain: http://infotech.nitk.ac.in</li></ul>"
  __global+=s
display(HTML(__global))
```
